```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the sample dataset
df1 = pd.read_csv("fact_spends(MitronBank).csv")

# Display the column names in the DataFrame
print(df1.columns)

df2 = pd.read_csv("dim_customers(MitronBank).csv")

# Display the column names in the second DataFrame
print(df2.columns)
# Merge the DataFrames on a common key (e.g., customer ID)
merged_df = pd.merge(df1, df2, on='customer_id', how='inner')

# Group Data by Age Group
spending_by_age_group = merged_df.groupby('age_group')['spend'].sum()

# Plot the Data
spending_by_age_group.plot(kind='bar', color='skyblue')
plt.title('Spending by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Spending')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python


# Load the sample dataset
df1 = pd.read_csv("fact_spends(MitronBank).csv")

# Display the column names in the DataFrame
print(df1.columns)

df2 = pd.read_csv("dim_customers(MitronBank).csv")

# Display the column names in the second DataFrame
print(df2.columns)
merged_df = pd.merge(df1, df2, on='customer_id', how='inner')

# Group Data by Marital Status
spending_by_marital_status = merged_df.groupby('marital status')['spend'].sum()

# Plot the Data
spending_by_marital_status.plot(kind='bar', color='skyblue')
plt.title('Spending by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Total Spending')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot the Data as a Pie Chart
plt.figure(figsize=(8, 6))
plt.pie(spending_by_marital_status, labels=spending_by_marital_status.index, autopct='%1.1f%%', startangle=140)
plt.title('Spending by Marital Status')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
# Generate colors for each marital status category
colors = {'Single': 'red', 'Married': 'green', 'Divorced': 'blue', 'Widowed': 'purple'}

# Plot the Data as a Scatter Plot
plt.figure(figsize=(10, 6))
for status, group in merged_df.groupby('marital status'):
    plt.scatter(group['spend'], group['marital status'], label=status, color=colors[status])

plt.title('Spending vs. Marital Status')
plt.xlabel('Spending')
plt.ylabel('Marital Status')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot the Data as a Box Plot
plt.figure(figsize=(10, 6))
plt.boxplot([merged_df[merged_df['marital status'] == status]['spend'] for status in merged_df['marital status'].unique()], 
            labels=merged_df['marital status'].unique())
plt.title('Spending Distribution by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Spending')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group Data by Occupation
spending_by_occupation = merged_df.groupby('occupation')['spend'].sum()

# Plot the Data as a Bar Plot
plt.figure(figsize=(10, 6))
spending_by_occupation.plot(kind='bar', color='skyblue')
plt.title('Spending by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Total Spending')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
